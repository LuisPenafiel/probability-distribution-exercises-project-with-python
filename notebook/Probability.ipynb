## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## **Normal Distribution**

```python
#Calculates the point estimate of the normal distribution
def dnorm(x, mean=0, sd=1):                 
                                      
    result = norm.pdf(x, loc=mean, scale=sd)
    return result

#Calculates the cumulative of the normal distribution
def pnorm(q, mean=0, sd=1):
  
    result = norm.cdf(q, loc=mean, scale=sd)
    return result

#Calculates the quantile function from the normal distribution
def qnorm(p, mean=0, sd=1):
  
    result = norm.ppf(p, loc=mean, scale=sd)
    return result

#Generates random variables from the normal distribution
def rnorm(n, mean=0, sd=1):

    result = norm.rvs(loc=mean, scale=sd, size=n)
    return result
```

```python
print(f"Probability that a Normal(10, 0.2) takes the value 2: {dnorm(2, mean=10, sd=0.2)}")
print(f"Probability that a Normal(10, 0.2) will take a value less than 2: {pnorm(2, mean=10, sd=0.2)}")
print(f"Which value of a Normal(10, 0.2) has a cumulative probability of 0.9? {qnorm(0.9, mean=10, sd=0.2)}")
print(f"Generate 2000 random values from a Normal(10, 0.2) distribution: {rnorm(2000, mean=10, sd=0.2)}")
```

```python
# Parameters for the normal distribution
mean = 0
sd = 1

# Generate an array of values for the x-axis
x = np.arange(-4, 4, 0.1)  # Range from -4 to 4 with a step of 0.1

                                           
pdf_values = dnorm(x, mean, sd)          # Calculate the probability density function (PDF) values for the x values
cdf_values = pnorm(x, mean, sd)          # Calculate the cumulative distribution function (CDF) values for the x values
       
quantiles = np.arange(0.01, 1, 0.01)     # Probabilities from 0.01 to 0.99 
ppf_values = qnorm(quantiles, mean, sd)  # Calculate quantile function (PPF) values for various probabilities,

# Generate random variables from the normal distribution
n = 1000  # Number of random samples
random_values = rnorm(n, mean, sd)


                            ###### Plotting ######

fig, axes = plt.subplots(2, 2, figsize=(12, 8))         # Create subplots to display all the visualizations

axes[0, 0].plot(x, pdf_values, 'r-', lw=2)              # Plot the PDF
axes[0, 0].set_title('Probability Density Function (PDF)')


axes[0, 1].plot(x, cdf_values, 'b-', lw=2)              # Plot the CDF
axes[0, 1].set_title('Cumulative Distribution Function (CDF)')


axes[1, 0].plot(quantiles, ppf_values, 'g-', lw=2)     # Plot the PPF
axes[1, 0].set_title('Quantile Function (PPF)')

axes[1, 1].hist(random_values, bins=20, density=True, alpha=0.5, color='purple')
axes[1, 1].set_title('Histogram of Random Variables')  # Plot the PPF

plt.tight_layout()
plt.show()                                            # Adjust layout and display the plots
```

**Uniform Distribution**

```python
def dunif(x, start=0, end=1):    # Calculates the point estimate of the uniform distribution

    result = uniform.pdf(x, loc=start, scale=end - start)
    return result

def punif(q, start=0, end=1):     #Calculates the cumulative of the uniform distribution

    result = uniform.cdf(q, loc=start, scale=end - start)
    return result

def qunif(p, start=0, end=1):     #Calculates the quantile function from the uniform distribution

    result = uniform.ppf(p, loc=start, scale=end - start)
    return result

def runif(n, start=0, end=1):    #Generates random variables from the uniform distribution

    result = uniform.rvs(loc=start, scale=end - start, size=n)
    return result
```

```python
print(f"Probability that a Uniform(10, 10.2) takes the value 2: {dunif(2, 10, 10.2)}")
print(f"Probability that a Uniform(10, 10.2) will take a value less than 2: {punif(2, 10, 10.2)}")
print(f"Which value of a Uniform(10, 10.2) has a cumulative probability of 0.9? {qunif(0.9, 10, 10.2)}")
print(f"Generate 2000 random values from a Uniform(10, 10.2) distribution: {runif(2000, 10, 10.2)}")
```

```python
# Parameters for the uniform distribution
start = 0
end = 1

x = np.arange(-0.5, 1.5, 0.01)             # Generate an array of values for the x-axis # Range from -0.5 to 1.5 with a step of 0.01

pdf_values = dunif(x, start, end)          # Calculate the probability density function (PDF) values for the x values
cdf_values = punif(x, start, end)          # Calculate the cumulative distribution function (CDF) values for the x values

quantiles = np.arange(0.01, 1, 0.01)       # Probabilities from 0.01 to 0.99
ppf_values = qunif(quantiles, start, end)  # Calculate quantile function (PPF) values for various probabilities


n = 1000  # Number of random samples
random_values = runif(n, start, end)       # Generate random variables from the uniform distribution

                            ###### Plotting ######

fig, axes = plt.subplots(2, 2, figsize=(12, 8))     # Create subplots to display all the visualizations

axes[0, 0].plot(x, pdf_values, 'r-', lw=2)          # Plot the PDF
axes[0, 0].set_title('Probability Density Function (PDF)')

axes[0, 1].plot(x, cdf_values, 'b-', lw=2)          # Plot the CDF
axes[0, 1].set_title('Cumulative Distribution Function (CDF)')

axes[1, 0].plot(quantiles, ppf_values, 'g-', lw=2)   # Plot the PPF
axes[1, 0].set_title('Quantile Function (PPF)')

axes[1, 1].hist(random_values, bins=20, density=True, alpha=0.5, color='purple')
axes[1, 1].set_title('Histogram of Random Variables') # Plot a histogram of random variables

plt.tight_layout()                                   # Adjust layout and display the plots
plt.show()                               
```
